# Predicting whether the car has had previous owners or is being sold by the first owner.


# importing libraries

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("cardekho_data.csv")
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["Car_Name"]=le.fit_transform(data["Car_Name"])
data["Fuel_Type"]=le.fit_transform(data["Fuel_Type"])
data["Seller_Type"]=le.fit_transform(data["Seller_Type"])
data["Transmission"]=le.fit_transform(data["Transmission"])
data

```

# extracting independent and dependent variable

```python
x=data.iloc[:,:-1].values
x
```

```python
y=data.iloc[:,-1].values
y
```


# splitting training and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

# fitting decision tree algorithm

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```
